```python
from pyspark.sql import SparkSession
spark = (SparkSession
         .builder
         .appName("IcebergDatamart")
         .getOrCreate()
        )
```

```python
table = spark.table("stage.spotify.playlists")
```

```python
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window
df_playlist = (table
      .select(
          col('playlist_id').alias('bk_id'),
          col('playlist_name').alias('name'),
          col('playlist_collaborative').alias('collaborative'),
          col('playlist_modified_at').alias('last_modified_at'),
          col('playlist_followers').alias('num_followers'),
          col('playlist_tracks').alias('num_edits')
      )
      .distinct()
      .withColumn("sk_id",row_number().over(Window.orderBy('bk_id')))
      .select(
          'sk_id',
          'bk_id',
          'name',
          'collaborative',
          'last_modified_at',
          'num_followers',
          'num_edits'
      )
#  .show()
)
df_playlist.show()
```

```python
(df_playlist
 .writeTo('gold.spotify_data_mart.d_playlist')
 .create()
)
```

```python
%%sql
select count(1) from gold.spotify_data_mart.d_playlist
```

```python
df_album = (table
      .select(
          col('track_album_uri').alias('bk_id'),
          col('track_album_name').alias('name')
      )
      .distinct()
      .withColumn("sk_id",row_number().over(Window.orderBy('bk_id')))
      .select(
          'sk_id',
          'bk_id',
          'name'
      )
#  .show()
)
df_album.show()
```

```python
(df_album
 .writeTo('gold.spotify_data_mart.d_album')
 .create()
)
```

```python
df_artist = (table
      .select(
          col('track_artist_uri').alias('bk_id'),
          col('track_artist_name').alias('name')
      )
      .distinct()
      .withColumn("sk_id",row_number().over(Window.orderBy('bk_id')))
      .select(
          'sk_id',
          'bk_id',
          'name'
      )
#  .show()
)
df_artist.show()
```

```python
(df_artist
 .writeTo('gold.spotify_data_mart.d_artist')
 .create()
)
```

```python
df_track = (table
      .select(
          col('track_uri').alias('bk_id'),
          col('track_name').alias('name'),
          col('track_duration').alias('duration')
      )
      .distinct()
      .withColumn("sk_id",row_number().over(Window.orderBy('bk_id')))
      .select(
          'sk_id',
          'bk_id',
          'name',
          'duration'
      )
#  .show()
)
df_track.show()
```

```python
(df_track
 .writeTo('gold.spotify_data_mart.d_track')
 .create()
)
```

```python
df_fact = (table
           .join(df_playlist, table.playlist_id == df_playlist.bk_id)
           .join(df_album, table.track_album_uri == df_album.bk_id)
           .join(df_artist, table.track_artist_uri == df_artist.bk_id)
           .join(df_track, table.track_uri == df_track.bk_id)
           .select(
               df_playlist.sk_id.alias('sk_playlist'),
               df_album.sk_id.alias('sk_album'),
               df_artist.sk_id.alias('sk_artist'),
               df_track.sk_id.alias('sk_track'),
               col('track_position')
           )
           .withColumn("id",
                       row_number().over(
                           Window.orderBy(
                               'sk_playlist', 
                               'sk_album', 
                               'sk_artist', 
                               'sk_track',
                               'track_position'
                           )
                       )
                      )
           .select(
               'id',
               'sk_playlist', 
               'sk_album', 
               'sk_artist', 
               'sk_track',
               'track_position'
           )
)
df_fact.show()
```

```python
(df_fact
 .writeTo('gold.spotify_data_mart.f_playlist_track')
 .create()
)
```

```python
%%sql
SELECT 
  f.id,
  dal.bk_id as album_id,
  dal.name as album_name,
  dar.bk_id as artist_id,
  dar.name as artist_name,
  dt.bk_id as track_id,
  dt.name as track_name,
  dt.duration as track_duration,
  dp.bk_id as playlist_id,
  dp.name as playlist_name,
  dp.num_followers as playlist_followers,
  dp.num_edits as playlist_edits,
  f.track_position
from gold.spotify_data_mart.f_playlist_track f 
  inner join gold.spotify_data_mart.d_album dal on f.sk_album = dal.sk_id 
  inner join gold.spotify_data_mart.d_artist dar on f.sk_artist = dar.sk_id
  inner join gold.spotify_data_mart.d_track dt on f.sk_track = dt.sk_id
  inner join gold.spotify_data_mart.d_playlist dp on f.sk_playlist = dp.sk_id
LIMIT 100;
```
